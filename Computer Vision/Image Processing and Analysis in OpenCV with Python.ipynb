```python
# All imports

import cv2
from matplotlib import pyplot as plt
%matplotlib inline

from pylab import rcParams
```

## Load Image

```python
picture_filename = "federernadal.jpg"
fedal = cv2.imread(picture_filename)
plt.imshow(fedal)
```

## Why is the image appearing blue?

Because BGR is the default OpenCV colour format. This means that blue and red planes get flipped.
Follow this answer on stackoverflow- https://stackoverflow.com/questions/42406338/why-cv2-imwrite-changes-the-color-of-pics
Let's check out solid colour palettes:

```python
col_pal = cv2.imread("colour_palette.jpg")
plt.imshow(col_pal)
```

```python
col_rgb = cv2.imread("rgbb.png")
plt.imshow(col_rgb)
```

These are not the true colours in the image. So let's view it with cv2.COLOR_BGR2RGB, i.e. correct colour pixels.

```python
pal_img_corrected = cv2.cvtColor(col_pal, cv2.COLOR_BGR2RGB)
plt.imshow(pal_img_corrected)
```

```python
rgb_img_corrected = cv2.cvtColor(col_rgb, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img_corrected)
```

Continuing with the tutorial...

```python
img_corrected = cv2.cvtColor(fedal, cv2.COLOR_BGR2RGB)
plt.imshow(img_corrected)
```

```python
plt.axis("off") #remove axes ticks
plt.imshow(img_corrected)
```

```python
rcParams['figure.figsize'] = 10, 12

plt.axis("off") #remove axes ticks
plt.imshow(img_corrected)
```

```python
gray_fedal = cv2.cvtColor(fedal, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_fedal, cmap = 'gray')
plt.axis("off") #remove axes ticks
plt.title('Grayscale Image')
```

```python
rcParams['figure.figsize'] = 10, 12
edges = cv2.Canny(img_corrected, 
                  threshold1=100, 
                  threshold2=200)

plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```

```python
rcParams['figure.figsize'] = 10, 12

edges = cv2.Canny(img_corrected, 
                  threshold1=500,   ## try different values here
                  threshold2=500) ## try different values here

plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```
